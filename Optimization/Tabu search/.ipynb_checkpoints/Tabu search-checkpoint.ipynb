```python
#Convert the file in a list and the list into a dictionary

def filetolist(file):
   openfile = open(file, "r")
   lines=openfile.readlines()
   list_=[]
   for line in lines:
       list_.append([int(s) for s in line.split() if s.isdigit() and int(s)>0])
   #create a list with the number of job in column one, time for M1 in column 2 and time for M2 in column 2
   for list2 in list_:
       list2.pop(0)
   print("LIST: ", list_)
   return list_

def listtodict(list_):
   dictionary = {}
   for i in range(len(list_)):
       dictionary[i+1]=list_[i]
   return dictionary
   print("DICTIONARY:", dictionary)
```

```python
#Obtain the first sequence with a Greedy algorithm
def firstsequence(list_):
    
    List1=[]
    List2=[]
    SumList1=[]
    SumDict1= {}
    SumList2=[]
    SumDict2= {}

    for i in range (len(list_)):
        if list_[i][0]>=list_[i][1]:
            List1.append(list_[i])
            SumList1.append(list_[i][0]+list_[i][1])
            SumDict1[i+1]=list_[i][0] + list_[i][1]
        else:
            List2.append(list_[i])
            SumList2.append(list_[i][0] + list_[i][1])
            SumDict2[i + 1] = list_[i][0] + list_[i][1]

    sorted_SumList1=sorted (SumList1)
    sorted_SumDict1 = dict(sorted (SumDict1.items(), key=operator.itemgetter(1)))
    sorted_SumList2=sorted (SumList2)
    sorted_SumDict2 = dict(sorted (SumDict2.items(), key=operator.itemgetter(1)))

    Sequence1 = list(sorted_SumDict1.keys())
    Sequence2 = list(sorted_SumDict2.keys())
    Sequence=[]

    for i in range(len(list_)):
        if i/2<len(Sequence1) and i/2<len(Sequence2):
            if i%2==0:
                Sequence.append(Sequence1[int(i/2)])
            else:
                Sequence.append(Sequence2[int((i-1)/2)])
        elif len(Sequence1)>len(Sequence2):
            Sequence.append(Sequence1[i-len(Sequence2)])
        else:
            Sequence.append(Sequence2[i-len(Sequence1)])
    return Sequence
```

```python
#Function to calculate the Completion Time

def CompletionTime(list_):
    completion_M1 = 0
    completion_total_M1 = []
    completion_M2 = 0
    completion_total_M2 = []
    for job in range(len(list_)):
        completion_M1 = completion_M1 + list_[job][0]
        completion_total_M1.append(completion_M1)
        if completion_M1 > completion_M2:
            max = completion_M1
        else:
            max = completion_M2
        completion_M2 = max + list_[job][1]
        completion_total_M2.append(completion_M2)
    return(sum(completion_total_M2))
```

```python
#Function to perform an insertion move

def oneinsertionmove(listtoinsert,originalindex,finalindex):
   if originalindex==finalindex:
       None #if you want to move the element of position one to position one, it is impossible
   elif originalindex<finalindex:
       listtoinsert.insert(finalindex+1,listtoinsert[originalindex])
       listtoinsert.pop(originalindex)
   else:
       listtoinsert.insert(finalindex, listtoinsert[originalindex])
       listtoinsert.pop(originalindex+1)
   return listtoinsert
```

```python
#Search for possible improvements applying insertion and swapping moves

def valuesofhybrid(Maxdistancetoswap,maxdistancetoinsert, listoriginal, Sequence):
   dictvalues = {}
   completiontime = CompletionTime(listoriginal)
   for j in range(1, Maxdistancetoswap + 1):
       for i in range(len(listoriginal) - j):
           templist = listoriginal.copy()
           tempseq=Sequence.copy()
           templist[i], templist[i + j] = templist[i + j], templist[i]
           deltatime = CompletionTime(templist) - completiontime
           if deltatime < 0:
               tempseq[i], tempseq[i + j] = tempseq[i + j], tempseq[i]
               dictvalues[i + 1, i + j + 1,1] = deltatime


         #  if len(dictvalues)>0:
        #       break
       #if len(dictvalues)>0:
          # break

   for j in range(2,maxdistancetoinsert+1):
        for i in range(len(listoriginal) - j):
           templist1 = listoriginal.copy()
           tempseq1=Sequence.copy()
           oneinsertionmove(templist1,i,i+j)
           oneinsertionmove(tempseq1, i, i + j)
           deltatime1 = CompletionTime(templist1) - completiontime
           templist2=listoriginal.copy()
           tempseq2 = Sequence.copy()
           oneinsertionmove(tempseq2, i+j, i)
           oneinsertionmove(templist2,i+j,i)
           deltatime2 = CompletionTime(templist2) - completiontime
           if deltatime2 < 0:
              dictvalues[i + j+1, i+1,2] = deltatime2
           if deltatime1<0:
               dictvalues[i+1,i+j+1,2]=deltatime1

           #if len(dictvalues) >1:
          #     break
        #if len(dictvalues) >1:
         #  break
   sorted_dictvalues = sorted(dictvalues.items(), key=operator.itemgetter(1), reverse=False)
   #print("Dictvalues ", sorted_dictvalues)
   #print("[0][1][0]", sorted_dictvalues[0][1][0])
   #print("SEq: ", sequences)
   #print("Reductions: ", reductions)
   return sorted_dictvalues
```

```python
#Apply the hybrid algorithm until a local optimum is found:

def hybrid(Maxdistancetoswap, maxdistancetoinsert, listoriginal, Sequence,optimumsfound,tabulist):
   It = 0
   list_to_change = listoriginal.copy()
   seq_to_change=Sequence.copy()
   completiontime = CompletionTime(list_to_change)
   #Maximum iterations allowed:2000. it can be changed
   #print("Tabu list: ", tabulist)
   while It < 2000:
       dictvalues = valuesofhybrid(Maxdistancetoswap, maxdistancetoinsert, list_to_change, seq_to_change)
       # Dictvalues is now a dictionary as indicated in the values of swaps and insertions
       while len(dictvalues)>0 and optimumsfound>0:
           newsequence = seq_to_change.copy()
           if dictvalues[0][0][2] == 1:
               newsequence[dictvalues[0][0][0] - 1], newsequence[dictvalues[0][0][1] - 1] = newsequence[dictvalues[0][0][1] - 1], newsequence[dictvalues[0][0][0] - 1]
           elif dictvalues[0][0][2] == 2:
               oneinsertionmove(newsequence, dictvalues[0][0][0] - 1, dictvalues[0][0][1] - 1)
           #print("New seq: ", newsequence)
           if newsequence not in tabulist:
               #print("Sequence not in tabu list ")
               break
           else:
               #print("SEQUENCE IN TABU LIST!!!.")
               dictvalues.pop(0)
               print("Tabu found")


       if dictvalues:
           It += 1
           #if len(dictvalues)>1 and dictvalues[0][1][0] == dictvalues[1][1][0] and dictvalues[0][0][2]==1 and dictvalues[1][0][2]==2:
            #   dictvalues.pop(0)
           if dictvalues[0][0][2]==1:

               list_to_change[dictvalues[0][0][0] - 1], list_to_change[dictvalues[0][0][1] - 1] = list_to_change[dictvalues[0][0][1] - 1], list_to_change[dictvalues[0][0][0] - 1]
               seq_to_change[dictvalues[0][0][0] - 1], seq_to_change[dictvalues[0][0][1] - 1] = seq_to_change[dictvalues[0][0][1] - 1], seq_to_change[dictvalues[0][0][0] - 1]
               #print("new seq after swap: ", seq_to_change)
               #print("NEWLIST AFTER NEIGHB:  ", list_to_change)
               completiontime = CompletionTime(list_to_change)
               #print("TOTAL COMPLETION TIME IS: ", completiontime)
               #print("Number of iteration: ", It)
               
           elif dictvalues[0][0][2]==2:

                # Do the insertion and the list is updated with that optimal insertion, update the sequence of jobs
                oneinsertionmove(list_to_change, dictvalues[0][0][0] - 1, dictvalues[0][0][1] - 1)
                oneinsertionmove(seq_to_change, dictvalues[0][0][0] - 1, dictvalues[0][0][1] - 1)
                #print("new seq after ins: ", seq_to_change)
                #print("NEWLIST AFTER NEIGHB:  ", list_to_change)
                completiontime = CompletionTime(list_to_change)
                #print("TOTAL COMPLETION TIME IS: ", completiontime)
                #print("Number of iteration: ", It)
                
           listtoappend=list_to_change.copy()
           listofjobs.append(listtoappend)
           seqtoappend = seq_to_change.copy()
           tabulist.append(seqtoappend)
           #print("Tabu list AFTER APPEND: ", tabulist)
       else:  # when dictvalues = None, we reached a local optimal
           break
   #print("Local optimal reached.")
   print(completiontime)
   #print("Total Iterations: ", It)
   return (list_to_change, seq_to_change, completiontime,tabulist)
```
