### Conteo de frecuencias de los lenguajes de programación

La columna que contiene información sobre los lenguajes de programación utilizados es `tools_programming_languages`. Sus valores son strings con los lenguajes seleccionados separados por comas.



```python
df.tools_programming_languages.head()
```

Las siguientes celdas de código separan estos lenguajes de programación y cuentan la frecuencia con la que aparecen.

No es necesario entender este código en profundidad, aunque sí es un buen ejercicio.

```python
# Convert the comma-separated string of languages to a list of string.
# Remove 'ninguno de los anteriores' option, spaces and training commas.
def split_languages(languages_str):
  if not isinstance(languages_str, str):
    return []
  # Remove 'other' option
  languages_str = languages_str.lower().replace('ninguno de los anteriores', '')
  # Split string into list of items
  # Remove spaces and commas for each item
  return [lang.strip().replace(',', '')
          for lang in languages_str.split()]

# Create a new column with the list of languages
df_analisis.loc[:, 'cured_programming_languages'] = df_analisis.tools_programming_languages\
    .apply(split_languages)
if 'cured_programming_languages' not in relevant_columns:
    relevant_columns.append('cured_programming_languages') 

# Duplicate each row of df for each programming language
# mentioned in the response.
# We only include in df_lang the columns we are going to analyze later, so we
# don't duplicate innecesary information.
df_lang = df_analisis.cured_programming_languages\
    .apply(pd.Series).stack()\
    .reset_index(level=-1, drop=True).to_frame()\
    .join(df_analisis[relevant_columns])\
    .rename(columns={0: 'programming_language'})
# Horrible programming style! But a lot of data science code can be written with
# as concatenations of functions (pipelines), and there's no elegant way of
# doing that on Python.
print(df_lang[:10])
print(df_lang.count())
```

En la columna `programming_language` se encuentra cada lenguaje por separado. Notar que si una respuesta contenía 3 lenguajes, como `"HTML, Javascript, Python"`, la fila ha sido replicada 3 veces. Por ello, hay tres filas con índice 1.

```python
language_count = df_lang.programming_language.value_counts().reset_index().rename(columns={'index': 'language', 'programming_language': 'frequency'})
lenguajes_interes = language_count[:11]['language'].tolist()

print(language_count[:12])
print()
print(lenguajes_interes)
```

Tomamos los 11 lenguajes más usados, ya que a partir del 12avo la frecuencia comienza a ser menos de la mitad del anterior.

Para analizar la independencia de las variables, preguntamos si  los programadores con título universitario tienen mayor probabilidad de ganar por encima de la media de todos los salarios. Para ello, primero filtramos la base por aquellos datos no nulos en el campo profile_studies_level

```python
df_analisis_studies = df_analisis.dropna(subset = ['profile_studies_level'])
avg_salary_studies = df_analisis_studies['salary_monthly_NETO'].mean()

univ_o_mas_lista=['Universitario','Posgrado/Especialización','Maestría','Doctorado','Posdoctorado']
universitarios_o_mas = df_analisis_studies['profile_studies_level'][df_analisis_studies.profile_studies_level.isin(univ_o_mas_lista)].count()

df_universitarios_o_mas = df_analisis_studies[df_analisis_studies.profile_studies_level.isin(univ_o_mas_lista)]

#Proporción de universitarios o mas
proporcion_univ_o_mas = universitarios_o_mas/df_analisis['profile_studies_level'].count()

print("Proporcion universitarios: ", round(proporcion_univ_o_mas,4))

#Proporción de programadores que ganan más que la media
above_avg = len(df_analisis_studies[df_analisis_studies['salary_monthly_NETO'] >= avg_salary_studies])
prob_above_avg = above_avg / len(df_analisis_studies)
print("Proporción más que la media: ",round(prob_above_avg,4))

#Proporción de programadores universitarios y ganan más que la media
above_avg_univ = len(df_universitarios_o_mas[df_universitarios_o_mas['salary_monthly_NETO'] >= avg_salary_studies])
prob_above_avg_univ = above_avg_univ / len(df_universitarios_o_mas)
print("Proporción más que la media dado que es universitario: ",round(prob_above_avg_univ,4))
```

A: probabilidad de ganar más que la media del salario neto

B: probabilidad de ser contar con estudios universitarios o superiores

Como la probabilidad de A es distinta a la probabilidad de A dado B, las variables no son independientes.
