###2.2 Imputar con KNN

Vemos la cantidad de datos:

````python
melb_df_encoded.describe()
````


Para la imputación de los datos mediante el metodo KNN no es necesario escalar o normalizar previamente los datos ya que la distancia relativa entre los puntos sigue siendo la misma, por lo que si escalaramos luego de la imputación, el resultado sería el mismo a si escalamos y luego imputamos.

````python
from sklearn.experimental import enable_iterative_imputer
from sklearn.neighbors import KNeighborsRegressor
from sklearn.impute import IterativeImputer

melb_data_mice = melb_df_encoded.copy(deep=True)

mice_imputer = IterativeImputer(random_state=0, estimator=KNeighborsRegressor())
melb_data_mice[['YearBuilt','BuildingArea']] = mice_imputer.fit_transform(
    melb_data_mice[['YearBuilt', 'BuildingArea']])
````

````python
melb_data_mice.describe()
````

###2.3 Gráfico

Gráfico comparando las distribuciones de datos obtenidas con cada método de imputación.

````python
mice_year_built = melb_data_mice.YearBuilt.to_frame()
mice_year_built['Imputation'] = 'KNN over YearBuilt'
melb_year_build = melb_df_encoded.YearBuilt.dropna().to_frame()
melb_year_build['Imputation'] = 'Original'
data = pd.concat([mice_year_built, melb_year_build])
fig = plt.figure(figsize=(8, 5))
g = seaborn.kdeplot(data=data, x='YearBuilt', hue='Imputation')
````

````python
mice_BuildingArea = melb_data_mice.BuildingArea.to_frame()
mice_BuildingArea['Imputation'] = 'KNN over BuildingArea'
melb_BuildingArea = melb_df_encoded.BuildingArea.dropna().to_frame()
melb_BuildingArea['Imputation'] = 'Original'
data = pd.concat([mice_BuildingArea, melb_BuildingArea])
fig = plt.figure(figsize=(8, 5))
g = seaborn.kdeplot(data=data, x='BuildingArea', hue='Imputation')
````

2. Grafique la varianza capturada por los primeros $n$ componentes principales, para cada $n$.
3. En base al gráfico, seleccione las primeras $m$ columnas de la matriz transformada para agregar como nuevas características al conjunto de datos.

###3.1 Aplicar PCA

Sí, es necesario normalizar antes de aplicar PCA.

Pero incluso antes, terminaremos de imputar las columnas y filas que aún presentan datos en blanco para poder aplicar PCA. Estas son zipcode,	airbnb_price_mean,	airbnb_record_count, y Car

````python
for column, data in melb_data_mice.iteritems():
    if data.count() < melb_data_mice.shape[0]:
        print(f"La columna {column} tiene menos de ", melb_data_mice.shape[0], " datos.")
````

Para la columna Car imputaremos la media en función del tipo de vivienda. Para las otras 3 columnas utilizaremos el mismo método que arriba KNN.

````python
melb_data_complete = melb_data_mice.copy(deep=True)
````

````python
# Calcular el promedio para cada tipo de vivienda
promedios = melb_data_mice[['Type_h', 'Type_t', 'Type_u', 'Car']].groupby(['Type_h', 'Type_t', 'Type_u'])['Car'].mean().reset_index()
promedios
````

````python
#Lo asignamos a 3 variables
prom_u,prom_t, prom_h = promedios['Car'][0],promedios['Car'][1],promedios['Car'][2]
prom_u,prom_t, prom_h
````

````python
#Y completamos el nuevo df melb_data_complete con esos valores para cada tipo de vivienda
for index, row in melb_data_complete.iterrows():
    if pd.isnull(row['Car']):
        if row['Type_u'] == 1:
            melb_data_complete.at[index, 'Car'] = prom_u
        elif row['Type_t'] == 1:
            melb_data_complete.at[index, 'Car'] = prom_t
        elif row['Type_h'] == 1:
            melb_data_complete.at[index, 'Car'] = prom_h

# Verificar los valores faltantes
print(melb_data_mice['Car'].isnull().sum())
print(melb_data_complete['Car'].isnull().sum())
````

````python
faltan_imputar = ['zipcode','airbnb_price_mean','airbnb_record_count']
melb_data_complete[faltan_imputar].count()
````

````python
mice_imputer_2 = IterativeImputer(random_state=0, estimator=KNeighborsRegressor())
melb_data_complete[faltan_imputar] = mice_imputer_2.fit_transform(
    melb_data_complete[faltan_imputar])

print("Después de imputar")
melb_data_complete[faltan_imputar].count()
````

````python
#Checkeamos que no queden columnas sin valores
for column, data in melb_data_complete.iteritems():
    if data.count() < melb_data_complete.shape[0]:
        print(f"La columna {column} tiene menos de ", melb_data_complete.shape[0], " datos.")
````

````python
#Definir n como se pide
n = min(20, melb_data_complete.shape[0])
n
````

````python
columnas_pca = ['PCA_' + str(i+1) for i in range(n)] 

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
# Normalizar los datos
scaler = StandardScaler()
melb_df_scaled = scaler.fit_transform(melb_data_complete)

# Aplicar PCA
pca = PCA(n_components=n)  # Especifica el número de componentes principales a obtener
components = pca.fit_transform(melb_df_scaled)

# Crear un nuevo DataFrame con los componentes principales
melb_df_pca = pd.DataFrame(data=components, columns= columnas_pca)

# Imprimir el DataFrame resultante
print(melb_df_pca)
````

###3.2 Graficar varianza

````python
import matplotlib.pyplot as plt

var_exp = pca.explained_variance_ratio_
cum_var_exp = numpy.cumsum(var_exp)

# Grafica la varianza explicada acumulada
plt.plot(range(1, len(cum_var_exp) + 1), cum_var_exp, marker='o')
plt.xlabel('Número de componentes principales')
plt.ylabel('Varianza explicada acumulada')
plt.title('Varianza explicada acumulada por componentes principales')
plt.show()
````

###3.3 Elegir m columas del PCA para agregar a la base de datos

````python
cum_var_exp
````

Dedidimos agregar las primeras 17 componentes principales de modo de poder explicar el 80% de la varianza

````python
df_pca_toadd =  melb_df_pca.drop(melb_df_pca.columns[-3:], axis=1)
print(df_pca_toadd[:3])
df_final = pd.concat([melb_data_complete, df_pca_toadd], axis=1)
print("\n Final dataframe:")
df_final[:3]
````

###4 Resolución

Ejemplo de juguete que venía ya hecho en el entregable:

````python
## Small example
from sklearn.decomposition import PCA
from sklearn.preprocessing import OneHotEncoder

## If we process our data with the following steps:
categorical_cols = ['Type']
numerical_cols = ['Rooms', 'Distance']
new_columns = []

# Step 1: encode categorical columns
encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
X_cat = encoder.fit_transform(melb_df[categorical_cols])
for col, col_values in zip(categorical_cols, encoder.categories_):
  for col_value in col_values:
    new_columns.append('{}={}'.format(col, col_value))
print("Matrix has shape {}, with columns: {}".format(X_cat.shape, new_columns))

# Step 2: Append the numerical columns
X = numpy.hstack([X_cat, melb_df[numerical_cols].values])
new_columns.extend(numerical_cols)
print("Matrix has shape {}, with columns: {}".format(X_cat.shape, new_columns))

# Step 3: Append some new features, like PCA
pca = PCA(n_components=2)
pca_dummy_features = pca.fit_transform(X)
X_pca = numpy.hstack([X, pca_dummy_features])
new_columns.extend(['pca1', 'pca2'])

## Re-build dataframe
processed_melb_df = pd.DataFrame(data=X_pca, columns=new_columns)
processed_melb_df.head()
````
