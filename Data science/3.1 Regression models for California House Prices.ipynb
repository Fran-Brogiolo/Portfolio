```python
def polinomial_regression(X_train, y_train, X_test, y_test, degree): #return error training, error test
    #degree = degree
    pf = PolynomialFeatures(degree)
    lr = LinearRegression()#fit_intercept=False)  # dejamos el fit_intercept = True, no sabemos si hay bias
    model = make_pipeline(pf, lr)
    model.fit(X_train, y_train)
    y_train_pred_pol = model.predict(X_train)
    y_test_pred_pol = model.predict(X_test)
    train_error_pol = mean_squared_error(y_train, y_train_pred_pol)
    test_error_pol = mean_squared_error(y_test, y_test_pred_pol)
    
    return train_error_pol,test_error_pol
```

```python
#Ej 7.
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error

degrees_ridge = [1,2,3]  # Grado del polinomio
alphas = [0.1,0.5, 1, 10, 100]  # Parámetro de regularización (ajusta según sea necesario)
```

```python
def ridge_regression(X_train_all, y_train, X_test_all, y_test, degree, alpha):
    poly_features = PolynomialFeatures(degree=degree)
    X_train_poly = poly_features.fit_transform(X_train_all)
    X_test_poly = poly_features.transform(X_test_all)

    ridge_model = make_pipeline(Ridge(alpha=alpha))
    ridge_model.fit(X_train_poly, y_train)

    y_train_pred = ridge_model.predict(X_train_poly)
    y_test_pred = ridge_model.predict(X_test_poly)

    train_error = mean_squared_error(y_train, y_train_pred)
    test_error = mean_squared_error(y_test, y_test_pred)
    #print("Polinomio grado ", degree, " alpha " , alpha)
    #print("Error de entrenamiento (MSE):", train_error)
    #print("Error de prueba (MSE):", test_error)
    return test_error
```

```python
import matplotlib.pyplot as plt

# Generar un gráfico por cada grado
for i, degree in enumerate(degrees_ridge):
    # Inicializar una lista para almacenar los errores de prueba por alpha
    test_errors = []

    # Calcular el error de prueba para cada alpha
    for alpha in alphas:
        test_error = ridge_regression(X_train_all, y_train, X_test_all, y_test, degree, alpha)
        test_errors.append(test_error)

    # Crear un nuevo gráfico para el grado actual
    plt.figure(figsize=(4, 4))
    plt.plot(alphas, test_errors, color='blue')
    plt.xlabel('Alpha')
    plt.ylabel('Test Error')
    plt.title(f'Test Error vs Alpha (Degree {degree})')
    plt.show()
```

No se observa ninguna mejora con respecto al modelo sin regularización.
